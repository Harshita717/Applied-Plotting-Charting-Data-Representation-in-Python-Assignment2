```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('IndiaWeather.csv')
```

```python
plt.figure()
df=df.where(df['YEAR']>2004).dropna()
df
```

```python
plt.figure()
df['maxtemp']=df[['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']].max(axis=1)
df['mintemp']=df[['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']].min(axis=1)
years=np.array([2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015])
plt.plot(years,df['maxtemp'],years,df['mintemp'])
plt.gca().fill_between(years,df['maxtemp'],df['mintemp'],facecolor='blue',alpha=0.25)
plt.title('Variation in temperature in India from 2005 to 2014')
plt.xlabel('Years')
plt.ylabel('Temperature')
plt.legend()
low=24.50
high=34.20
ifor=0
for i in df.index:
    if df.loc[i]['maxtemp']>high:
        df.at[i, 'hot'] = df.loc[i]['maxtemp']
    else:
         df.at[i, 'hot']='Nan'
        
for i in df.index:
    if df.loc[i]['mintemp']<low:
        df.at[i, 'cold'] = df.loc[i]['mintemp']
    else:
         df.at[i, 'cold']='Nan'

plt.scatter(years,df['hot'],color='red')
plt.scatter(years,df['cold'],color='blue')
plt.legend()
df
```
